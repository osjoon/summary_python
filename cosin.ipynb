```python
from konlpy.tag import Kkma
```

```python
kkma = Kkma()
microbes=[]
total_list = []
for x in tqdm(range(0,len(df.index))):
    try:
        i = df.iloc[x]
        species = i['species'].split()
        limit = i['limit'].split()
        use = i['use'].split()
        colony_color = i['colony_color'].split()
        pathogen = i['pathogen']
        

        nouns_description_result = kkma.nouns(pathogen)

        total_list.extend(limit)
        total_list.extend(use)
        total_list.extend(pathogen)
        total_list.extend(colony_color)
        total_list.extend(nouns_description_result)

        
        microbes.append([species,total_list])
    except:
        pass
```

```python
df['okt']= df['limit'] + ',' + df['use'] + ',' + df['pathogen'] +  ',' + df['colony_col']
```

```python
from konlpy.tag import Okt

okt=Okt()

# 토큰화
total=[]
for i in tqdm(df['okt']):
    try:
        print(i)
        t=okt.nouns(i)
        total.append(" ".join(t))

    except:
        pass
```

```python
len(total)
```

```python
df['okt']=total

df.tail()
```

```python
# 벡터화
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer()
test=df["okt"].to_list()
dtm=cv.fit_transform(test)
dtm=pd.DataFrame(dtm.toarray(),columns=cv.get_feature_names_out())
print(cv.get_feature_names_out())
dtm
```

```python
import numpy as np
# 코사인 유사도 : 두 벡터 사이 각도의 코사인 값 (-1 ~ 1)
def make_cosine(a,b):
    return np.dot(a,b) / (np.linalg.norm(a) * np.linalg.norm(b)) 
```

```python
# 테스트 값
text = '치즈에 파란 곰팡이가 피었다.'
text = okt.nouns(text)
text = " ".join(text)

dtm_2=cv.transform([text])
target=dtm_2.toarray()[0]
```

```python
box=[]
for i in range(len(dtm)):
    sample=dtm.iloc[i].to_numpy()
    cosine=make_cosine(target,sample)
    box.append(cosine)
 
# 데이터프레임에 유사도 열 추가
df['유사도분석']=box

# 유사도에 따라 데이터프레임 정렬
df.sort_values('유사도분석',ascending=False,inplace=True)
df.head(10)
```
